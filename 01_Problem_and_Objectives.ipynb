In subsurface geology, correctly identifying **rock facies** from well log measurements is important for:

- Reservoir characterization

- Oil & gas exploration

- Understanding lithology distribution

## 👉 Problem:

Given well log measurements (like Gamma Ray, Resistivity, Porosity, etc.), can we **predict the facies class** of each depth interval?

- This is a **multi-class classification problem**.

## Problem Statement

Lithofacies classification plays a critical role in reservoir characterization, hydrocarbon exploration, and subsurface geological interpretation. Traditional lithofacies identification methods often rely heavily on manual interpretation of well log data, which can be time-consuming, subjective, and challenging in heterogeneous reservoir systems.

In complex geological environments, overlapping petrophysical properties and transitional lithological boundaries may reduce the efficiency and consistency of conventional interpretation techniques. As a result, there is increasing interest in the application of machine learning techniques for automated and intelligent lithofacies prediction.

This study investigates the application of machine learning algorithms to classify lithofacies using well log data. The research focuses on improving predictive accuracy, understanding feature contributions, and enhancing model interpretability for reliable subsurface characterization.

The target variable consists of nine lithofacies classes:

| 9 | Phylloid-Algal Bafflestone |

### Geological Significance

- **Sandstone**: Typically associated with high reservoir quality due to favorable porosity and permeability.
- **Coarse Siltstone**: Transitional facies with moderate reservoir potential.
- **Fine Siltstone**: Fine-grained sedimentary rock with lower permeability.
- **Marine Siltstone and Shale**: Deposited in marine environments and often act as seals.
- **Mudstone**: Very fine-grained rock with generally poor reservoir properties.
- **Wackestone**: Carbonate facies containing mud-supported grains.
- **Dolomite**: Carbonate rock that may exhibit enhanced porosity through diagenetic processes.
- **Packstone-Grainstone**: Grain-supported carbonate facies with good reservoir potential.
- **Phylloid-Algal Bafflestone**: Reef-associated carbonate facies formed by algal framework structures.
