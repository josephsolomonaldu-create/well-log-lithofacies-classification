# 📌 Step 01: Problem & Objectives

In subsurface geology, correctly identifying **rock facies** from well log measurements is important for:

- Reservoir characterization

- Oil & gas exploration

- Understanding lithology distribution

## 👉 Problem:

Given well log measurements (like Gamma Ray, Resistivity, Porosity, etc.), can we **predict the facies class** of each depth interval?

- This is a **multi-class classification problem**.

## 🎯 Objectives

Build a machine learning model that:

**1**. Takes well log features as input

**2**. Predicts the correct facies label

**3**. Maximizes classification performance (Accuracy + F1-score)

| Column     | Meaning                                     | Type                     |
| ---------- | ------------------------------------------- | ------------------------ |
| **Facies** | Rock type (TARGET)                          | categorical              |
| Formation  | Geological formation name                   | categorical (drop later) |
| Well Name  | Well identifier                             | categorical (drop later) |
| Depth      | Depth in the well                           | numeric (not predictive) |
| GR         | Gamma Ray → shale indicator                 | numeric                  |
| ILD_log10  | Resistivity (log scale) → fluid & lithology | numeric                  |
| DeltaPHI   | Porosity difference                         | numeric                  |
| PHIND      | Porosity index                              | numeric                  |
| PE         | Photoelectric effect → mineral type         | numeric                  |
| NM_M       | Marine indicator                            | numeric                  |
| RELPOS     | Relative position in formation              | numeric                  |
